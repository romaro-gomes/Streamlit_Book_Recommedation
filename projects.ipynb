# Import ratings dataset

```python
ratings = pd.read_csv('data/Ratings.csv')
ratings.head()
```

# Rename Datasets columns

```python
books.rename(columns={
    'ISBN':'isbn',
    'Book-Title':'title',
    'Book-Author':'author',
    'Year-Of-Publication':'year',
    'Publisher':'publisher',
    'Image-URL-L':'img_url'},
    inplace=True
)

user.rename(columns={
    'User-ID':'user_id'},
    inplace=True
)

ratings.rename(columns={
    'User-ID':'user_id',
    'ISBN':'isbn',
    'ISBN':'isbn',
    'Book-Rating':'ratings'},
    inplace=True
)
```

# Filtring users with more reads

```python
#ratings['user_id'].value_counts()

x=ratings['user_id'].value_counts() > 100
# x
x[x].shape #show only who is equal True
```

```python
y = x[x].index
ratings = ratings[ratings['user_id'].isin(y)]
ratings
```

```python
ratings.shape
```

# Join Datasets

```python
ratings_with_books = ratings.merge(books, on='isbn')
```

```python
ratings_with_books.head(2)
```

```python
ratings_with_books.shape
```

# Number of ratings 

```python
num_ratings= ratings_with_books.groupby('title')['ratings'].count().reset_index()
```

```python
num_ratings.head(2)
```

```python
num_ratings.rename(columns={'ratings':'num_of_ratings'}, inplace=True)
```

```python
num_ratings.head(2)
```

# Filter books with 10 readers or more


```python
final_ratings = ratings_with_books.merge(num_ratings, on='title')
```

```python
final_ratings
```

```python
final_ratings.shape
```

```python
final_ratings=final_ratings[final_ratings['num_of_ratings']>=50]
final_ratings.shape
```

```python
final_ratings.drop_duplicates(['user_id','title'],inplace=True)
final_ratings.shape
```

# Pivot book table


```python
pivot_book=final_ratings.pivot_table(columns='user_id', index='title',values='ratings')
```

```python
pivot_book
```

```python
pivot_book.shape
```

```python
pivot_book.fillna(0,inplace=True)
```

```python
pivot_book
```

# Dataset with many zeros

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse= csr_matrix(pivot_book)
```

```python
book_sparse
```

# Create model

```python
from sklearn.neighbors import NearestNeighbors
model= NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(pivot_book.iloc[237,:].values.reshape(1,-1), n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(pivot_book.index[suggestion[i]])
```

```python
book_name = pivot_book.index
book_name
```

# Save the Model

```python
import pickle
import os
if not os.path.exists('artifacts'):
   os.makedirs('artifacts')
   
pickle.dump(model,open('artifacts/model.pkl','wb'))
pickle.dump(book_name,open('artifacts/books_name.pkl','wb'))
pickle.dump(final_ratings,open('artifacts/final_ranting.pkl','wb'))
pickle.dump(pivot_book,open('artifacts/book_pivot.pkl','wb'))
```
